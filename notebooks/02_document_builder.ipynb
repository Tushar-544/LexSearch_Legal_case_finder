# 02 — Document Builder
Converts each row of `cases_clean.csv` into a structured JSON document and saves `documents.jsonl`.

```python
import pandas as pd
import json
from pathlib import Path

CLEAN_PATH = Path('data/processed/cases_clean.csv')
OUT_PATH   = Path('data/processed/documents.jsonl')
OUT_PATH.parent.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(CLEAN_PATH, low_memory=False)
print(f'Loaded {len(df):,} rows')
```

```python
def row_to_doc(row):
    """Convert a DataFrame row into a structured document dict."""
    return {
        'doc_id'       : str(row['case_id']),
        'case_id'      : str(row['case_id']),
        'case_no'      : str(row.get('case_no', '')),
        'petitioner'   : str(row.get('petitioner', '')),
        'respondent'   : str(row.get('respondent', '')),
        'judge'        : str(row.get('judge', '')),
        'bench'        : str(row.get('bench', '')),
        'decision_date': str(row.get('decision_date', '')),
        'year'         : int(row['year']) if pd.notna(row.get('year')) else None,
        'case_text'    : str(row.get('case_text', '')),
        'pdf_url'      : str(row.get('pdf_path', '')),
        'judgement_type': str(row.get('judgement_type', '')),
    }

docs = [row_to_doc(row) for _, row in df.iterrows()]
print(f'Built {len(docs):,} documents')
print('Sample:', json.dumps(docs[0], indent=2))
```

```python
with open(OUT_PATH, 'w', encoding='utf-8') as f:
    for doc in docs:
        f.write(json.dumps(doc, ensure_ascii=False) + '\n')

print(f'Saved → {OUT_PATH}')
```
